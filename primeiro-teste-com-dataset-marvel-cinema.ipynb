```python
#Inserindo nova coluna com o calculo de horas de cada filme no dataset
for index, row in cinema.iterrows():
    if(row['movie_duration'] != 0):
        total_horas = (row['movie_duration']/60)
        decimal_horas =  float(f'{total_horas:.2}') - int(total_horas)
        inteiro_total_horas = int(total_horas)
        decimal_horas = int(decimal_horas*60)
       
        if(decimal_horas < 10):
            HorasTotal = ('0' + str(inteiro_total_horas) + ':' + '0' + str(decimal_horas))
       
        elif(decimal_horas >= 10):     
            HorasTotal = '0' + str(inteiro_total_horas) + ':' + str(decimal_horas)
        
        
        cinema.loc[index, 'movie_hours'] = HorasTotal

    else:
        cinema.loc[index, 'movie_hours'] = 0

cinema

        
```
